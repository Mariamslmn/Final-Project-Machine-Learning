```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#create a new dataframe having the accuracy of every model applied on 2 datasets:

scores_df= pd.DataFrame({"Data Studied":['forest fires', 'refugees'],
                     "Linear regression":[-0.013,0.888],
                     "Logistic Regression":[0.515,0.423],
                     "Lasso Regression":[-0.006,0.817],
                     "Decision Tree":[0.531,0.538],
                     "Naive Bayes": [0.485,0.385],
                     "Random Forest":[0.59,0.5],
                     "Support Vector Machine":[0.515,0.5],
                     "KNN":[0.554,0.5],
                     "Neural Networks":[0.6,0.55],
                     "Elastic Net":[0.0129,0.717],
                     "K-Means":[-427120.866,-1474133.016]}
                     )
scores_df
```

```python
#set Data Studied as index
df2=scores_df.set_index('Data Studied')
df2.index.name = ''
df2
```

```python
#Transposing the DataFrame
df2 = df2.T
```

```python
#drop Kmeans from the dataset
df2=df2.drop(['K-Means'])

df2
```

```python
#reset the index to get the list of models
df2 = df2.reset_index(drop=False)
df2=df2.rename(index=str, columns={"index": "Models"})
df2
```

```python
#create a list for labels
data_list =  df2["Models"]

#Visualize: forest fires vs refugee predictions :
fig = plt.figure(figsize=(20, 8))
forest= df2['forest fires']
ref_study= df2['refugees']
x_axis = np.arange(0, len(df2))
tick_locations = []
for x in x_axis:
    tick_locations.append(x)



score_plt = plt.bar(x_axis,  forest, width=1.0, label='Forest Fires', color='lightskyblue', edgecolor="black", linewidth="0.75", align="center",alpha=0.5)

score2_plt = plt.bar(x_axis,  ref_study, width=1.0, label='Refugees', color='coral', edgecolor="black", linewidth="0.75", align="center",alpha=0.5)
plt.title(f"Models Accuracy for Forest Fires Dataset vs Refugees Dataset")

# Set our legend 
lgnd = plt.legend(handles=[score_plt, score2_plt], loc='upper left')

plt.xticks(tick_locations, data_list )

plt.savefig("../scores_bar.png")

plt.show()
```
